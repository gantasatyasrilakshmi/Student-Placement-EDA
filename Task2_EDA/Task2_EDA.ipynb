# Data Visualization 

# Placement Status Count

```python
plt.figure(figsize=(6,4))
sns.countplot(x="placement_status", data=df)
plt.title("Placement Status Distribution")
plt.xlabel("Placement Status")
plt.ylabel("Number of Students")
plt.show()
```

# Number of Students in Each Branch

```python
plt.figure(figsize=(10,5))
sns.countplot(x="branch", data=df)
plt.xticks(rotation=45)
plt.xlabel("Branch")
plt.ylabel("Number of Students")
plt.title("Students by Branch")
plt.show()
```

# Distribution of Student CGPA

```python
plt.figure(figsize=(8,5))
sns.histplot(df["cgpa"], bins=20, kde=True)
plt.xlabel("CGPA")
plt.ylabel("Number of Students")
plt.title("CGPA Distribution")
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.xlabel("Features")
plt.ylabel("Features")
plt.title("Correlation Heatmap")
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=df["cgpa"])
plt.xlabel("CGPA")
plt.ylabel("CGPA")
plt.title("CGPA Boxplot")
plt.show()
```

# College Tier Distribution

```python
sns.countplot(x='college_tier', data=df)
plt.xlabel("College Tier")
plt.ylabel("Number of Students")
plt.title("College Tier Distribution")
```

# Internships Distribution

```python
sns.countplot(x='internships', data=df)
plt.xlabel("Number of Internships")
plt.ylabel("Number of Students")
plt.title("Internships Distribution")
```

# Certifications Distribution

```python
sns.countplot(x='certifications', data=df)
plt.xlabel("Number of Certifications")
plt.ylabel("Number of Students")
plt.title("Certifications Distribution")
```

# Hackathons Distribution

```python
sns.countplot(x='hackathons', data=df)
plt.xlabel("Number of Hackathons")
plt.ylabel("Number of Students")
plt.title("hackathons Distribution")
```

# Backlogs Distribution

```python
sns.countplot(x='backlogs', data=df)
plt.xlabel("Number of Backlogs")
plt.ylabel("Number of Students")
plt.title("backlogs Distribution")
```

# Relationship graphs

# CGPA vs Salary


```python
sns.scatterplot(x='cgpa',
                y='salary_package_lpa',
                data=df)
plt.xlabel("CGPA")
plt.ylabel("Salary Package (LPA)")
```

# Coding Skills vs Salary

```python
sns.scatterplot(
x='coding_skills',
y='salary_package_lpa',
data=df)
plt.xlabel("Coding Skills")
plt.ylabel("Salary Package (LPA)")
```

# DSA vs Salary

```python
sns.scatterplot(
x='dsa_score',
y='salary_package_lpa',
data=df)
plt.xlabel("DSA Score")
plt.ylabel("Salary Package (LPA)")
```

# Communication Skills vs Salary

```python
plt.figure(figsize=(8,5))
plt.scatter(df["communication_skills"], df["salary_package_lpa"], alpha=0.3)

plt.title("Communication Skills vs Salary")
plt.xlabel("Communication Skills Score")
plt.ylabel("Salary Package (LPA)")

plt.show()
```

# Internships vs Salary

```python
plt.figure(figsize=(8,5))
plt.scatter(df["internships"], df["salary_package_lpa"], alpha=0.3)

plt.title("Internships vs Salary")
plt.xlabel("Number of Internships")
plt.ylabel("Salary Package (LPA)")

plt.show()
```

# Conclusion

Key Findings

• Dataset contains 100,000 student records and 18 features.

• Salary information is available only for placed students.

• Most students belong to Tier-2 and Tier-3 colleges.

• Students with higher CGPA generally receive better salary packages.

• Better coding skills and DSA scores are associated with higher salaries.

• Students with internships and certifications tend to have better placement outcomes.

• No duplicate records were found.

• Missing values exist only in Salary Package, which is expected because unplaced students do not receive salary offers.
